## 카드 뽑기 시뮬레이션

```python
import random as r
import matplotlib.pyplot as plt

draw = []   # 2장씩 묶어서 저장할 리스트
aa_bar = []
na_bar = []
cards = [
    "A", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X",
    "A", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X",
    "A", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X",
    "A", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X", "X",
    "X", "X"
]

# 카드 섞기
for _ in range(30):
    r.shuffle(cards)

# 500번 반복해서 2장씩 뽑기 (복원추출)
for i in range(1, 501):
    pair = []
    for x in range(2):
        a= 0
        a = r.choice(cards)
        pair.append(a)
        print("당신이 뽑은 카드: {}".format(a))
    print("{}번 ==============================".format(i))   # 카드 1장 뽑기
    draw.append(pair)                  # 2장 묶어서 저장

for pair in draw:
    if pair == ["A", "A"]:
      aa_bar.append(1)
      na_bar.append(0)
    else:
      aa_bar.append(0)
      na_bar.append(1)

# AA면 1, 아니면 0으로 변환

# 막대그래프 그리기
plt.figure(figsize=(12, 4))
plt.bar(range(len(draw)), aa_bar)
plt.title("A posision")
plt.xlabel("1~500")
plt.ylabel("AA (1 = appeared)")
plt.show()

# plt.figure(figsize=(12, 4))
# plt.bar(range(len(draw)), na_bar)
# plt.title("NA posision")
# plt.xlabel("1~500")
# plt.ylabel("NA (1 = appeared)")
# plt.show()

# AA 나온 위치 출력
# aa_index = [i for i, v in enumerate(aa_bar) if v == 1]
# print("AA 나온 위치:", aa_index)
```
